```python
run_event_stage_df_list[(run_event_stage_df_list['subject']=='DEV049') &
                        (run_event_stage_df_list['wave']==1) & 
                       (run_event_stage_df_list['stage']=='food')]
```

```python
run_event_stage_df_list[(run_event_stage_df_list['subject']=='DEV049') &
                        (run_event_stage_df_list['wave']==1) & 
                       (run_event_stage_df_list['stage']=='bid_prechoice')]
```

```python
food_plus_prechoice = pd.merge(
    run_event_stage_df_list[(run_event_stage_df_list['subject']=='DEV049') &
                        (run_event_stage_df_list['wave']==1) & 
                       (run_event_stage_df_list['stage']=='food')],
    run_event_stage_df_list[(run_event_stage_df_list['subject']=='DEV049') &
                        (run_event_stage_df_list['wave']==1) & 
                       (run_event_stage_df_list['stage']=='bid_prechoice')],
    on=['event_id','run']
)

food_plus_prechoice['total_duration'] = food_plus_prechoice.duration_x + food_plus_prechoice.duration_y
```
