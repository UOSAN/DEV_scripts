## six-second tasks results

```python
metadata = pickle.load(open("../data/first_subs_nifti_metadata.pkl",'rb'))
```

```python
loocv_results = pickle.load(open("../data/cv_train_test_ns_6s60subjs_sets_outer_n_loocv_original_response_format.pkl",'rb'))
print(len(loocv_results))
test_scores_same=loocv_results[0]
tt_results=loocv_results[1]
results_by_trainset_item=loocv_results[2]
```

```python
results = pd.concat([metadata,results_by_trainset_item],axis=1)

results['result_group'] = results.group + "_" + results.run
```

```python
within_group_predictions = within_group_performance(results,pearsonr_cor_func,'result_group') 
r2_score_within_subj = math.pow(np.mean(within_group_predictions),2)
print("Pearson's R^2 within subject and run:")
print(r2_score_within_subj)

within_subj_predictions = within_group_performance(results,pearsonr_cor_func,'group') 
r2_score_within_subj_run = math.pow(np.mean(within_subj_predictions),2)
print("Pearson's R^2 within subject only:")
print(r2_score_within_subj_run)



print("Pearson's R^2 across subjects:")
math.pow(pearsonr(results['y'],results['y_pred'])[0],2)
```

That's a lot better; it's legit-good, in fact...but it's not even masked, and I am worried we're getting too much finger-press signal built into this.

That's interesting--the within-subjects score is substantially better than the between-subjects score. So we are getting there. But it might still be finger pressing going on.

```python
within_subj_predictions_spearman = within_group_performance(results,spearmanr_cor_func,'result_group')
    
rho_score_within_subj = np.mean(within_subj_predictions)

print(rho_score_within_subj)
print(spearmanr(results['y'],results['y_pred'])[0])
```

## Response-normalized data

```python
loocv_results = pickle.load(open("../data/cv_train_test_ns_6s60subjs_sets_outer_n_loocv_response_normalized.pkl",'rb'))
print(len(loocv_results))
test_scores_same=loocv_results[0]
tt_results=loocv_results[1]
results_by_trainset_item=loocv_results[2]
weightmap = loocv_results[3]
```

```python
within_subj_predictions = within_group_performance(results,pearsonr_cor_func,'group') 
r2_score_within_subj_run = math.pow(np.mean(within_subj_predictions),2)
print("Pearson's R^2 within subject only:")
print(r2_score_within_subj_run)

within_group_predictions = within_group_performance(results,pearsonr_cor_func,'result_group') 
r2_score_within_subj = math.pow(np.mean(within_group_predictions),2)
print("Pearson's R^2 within subject and run:")
print(r2_score_within_subj)

print("Pearson's R^2 across subjects:")
math.pow(pearsonr(results['y'],results['y_pred'])[0],2)
```
