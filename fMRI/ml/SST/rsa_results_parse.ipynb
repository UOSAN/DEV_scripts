```python
import pandas as pd
import pickle
import numpy as np

data_path = "/gpfs/projects/sanlab/shared/DEV/nonbids_data/fMRI/ml/SST"
```

```python
correlations = pickle.load(open(data_path + '/correlations.pkl','rb'))
```

Let's get averages over the 15 subjects.

```python
correlations['anterior MPFC']['correct-stop inter minus intra'] = correlations['anterior MPFC']['correct-go, correct-stop']-correlations['anterior MPFC']['correct-stop']
```

```python
for region in correlations.keys():
    correlations[region]['correct-stop inter minus intra'] = correlations[region]['correct-go, correct-stop']-correlations[region]['correct-stop']
```

```python
overall_inter_group_similarity_difference = []
for region in correlations.keys():
    diff = np.mean(correlations[region]['correct-stop inter minus intra'])
    print(region + ': ' + str(diff))
    overall_inter_group_similarity_difference.append((region,diff))
```
